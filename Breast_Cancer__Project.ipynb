```python
df = pd.read_csv('data.csv')
df.head()

```

```python
# Check for empty values in every column
print(df.isnull().sum())

# Get a summary of the numbers (Mean, Min, Max)
df.describe()
```

```python
# Drop the empty column and the ID column
df = df.drop(columns=['Unnamed: 32', 'id'])
# Check the new size of data
print("New Data Shape:", df.shape)
```

```python
# Convert 'M' to 1 and 'B' to 0
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])

# Show the new table
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df)
plt.title('Count of Cancer Cases')
plt.show()
```

```python
plt.figure(figsize=(20, 18))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Separate the Features (X) from the Target (Y)
X = df.drop(columns='diagnosis')
Y = df['diagnosis']

# Split the data: 80% for Training, 20% for Testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

```python
# Import the Model
from sklearn.linear_model import LogisticRegression

# Create the AI Brain
model = LogisticRegression(max_iter=3000)

# Teach the Brain using the Training Data
model.fit(X_train, Y_train)

print("Model Training Completed Successfully!")
```

```python
from sklearn.metrics import accuracy_score

# Check accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on Training data : ', training_data_accuracy)

# Check accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on Test data : ', test_data_accuracy)
```
